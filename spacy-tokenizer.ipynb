Meaning:
doc is an instance of the class spacy.tokens.doc.Doc, or simply spacy.tokens.Doc.

The Doc class is spaCy's main container for an entire processed text.

It holds all tokens, spans, and annotations (like part-of-speech tags, named entities, etc.).

So, type(doc) is used to:

Identify what type of object doc is.

Help you understand which class you're working with, so you can look up its methods and properties.

```python
type(token)
```

Meaning:
The token object is an instance of the Token class from spacy.tokens.token.

This class represents a single word (or symbol) in the processed text.

So:

type(doc) → <class 'spacy.tokens.doc.Doc'> → whole document

type(token) → <class 'spacy.tokens.token.Token'> → single token (word/punctuat)

1. Use the right NLP model for your task.
If you want built-in features like detecting numbers, emails, or currency, use a full model like en_core_web_sm. Blank models like spacy.blank("bn") only do basic tokenization.

2. Never write for token.i in doc: — it's invalid.
token.i is a read-only attribute. Use for token in doc: and access the index with token.i.

3. You must give a single string to nlp(), not a list of lines.
Always use text = "".join(lines) or text = file.read() before passing to nlp(text).

4. token.like_email, token.like_num, and token.is_currency won't work with blank models.
These attributes only return useful results with full pipelines like English (en_core_web_sm). They don’t work with spacy.blank("bn").

5. Use regular expressions when spaCy doesn’t support a feature (like Bangla numbers or emails).
Regex is your best friend for extracting Bangla digits, time formats (like ১০:৩০), and emails from mixed-language text.

6. Use token.i to access the position of a token inside the doc.
If you want the token before a currency symbol, use doc[token.i - 1].

7. Don’t mix up variable names.
If you assign textdoc = nlp(text), then loop using for token in doc:, it will crash because doc is undefined. Keep naming consistent.

8. Use UTF-8 encoding when reading Bangla or multilingual files.
This avoids character misreading or encoding errors.

9. To extract things like “500 €”, first check if the token is currency, then get the previous token as the number.

10. Use token.text to access the word, and token.is_digit to check if it’s a digit (for English digits only).
Bangla digits (like ১৫০০) need regex, because spaCy doesn’t recognize them as digits by default.
